```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Plagiarism Detector/plagiarism_dataset.csv'

df = pd.read_csv(file_path)
```

```python
df
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
values = df['Task'].value_counts()
plt.bar(values.index, values.values)
plt.title('Number of submissions for each tasks')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
values = df.groupby('Task')['Category'].value_counts()
values.unstack().plot(kind='bar', stacked=True)
plt.title('Plagiarism Category for each Task')
plt.xlabel('Task')
plt.ylabel('Count')
plt.show()
```

```python
values = df['Category'].value_counts()
plt.pie(values, labels=values.index, autopct='%1.1f%%')
plt.title('Plagiarism Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
non_count = df[df['Category'] == 'non']
values = non_count['Task'].value_counts()
plt.bar(values.index, values.values)
plt.title('Plagiarism Category: Non')
plt.xlabel('Task')
plt.ylabel('Count')
plt.show()
```

```python
non_count = df[df['Category'] == 'cut']
values = non_count['Task'].value_counts()
plt.bar(values.index, values.values, color = 'c')
plt.title('Plagiarism Category: cut')
plt.xlabel('Task')
plt.ylabel('Count')
plt.show()
```

```python
non_count = df[df['Category'] == 'light']
values = non_count['Task'].value_counts()
plt.bar(values.index, values.values,color='g')
plt.title('Plagiarism Category: Light')
plt.xlabel('Task')
plt.ylabel('Count')
plt.show()
```

```python
non_count = df[df['Category'] == 'heavy']
values = non_count['Task'].value_counts()
plt.bar(values.index, values.values,color='r')
plt.title('Plagiarism Category: Heavy')
plt.xlabel('Task')
plt.ylabel('Count')
plt.show()
```
